# Fixed-Point Arithmetic with the spfpm Package
This notebook demonstrates how to use the `spfpm` FixedPoint package in Python for fixed-point arithmetic operations.

```python
## Package Installation
## The spfpm package was installed using pip in a virtual environment.
!pip show spfpm
```

```python
from FixedPoint import FXfamily, FXnum
```

## 1. Basic Fixed-Point Division
In this example, I calculate the value of 22/7 and compare it with 3.1415.

```python
x = FXnum(22) / FXnum(7)
y = FXnum(3.1415)

print("x =", x)
print("y =", y)
print("x - y =", x - y)
```

## 2. Using a Fixed Precision Family
Here I define a fixed-point family with 16 fractional bits and perform arithmetic operations.

```python
fam16 = FXfamily(16)
a = FXnum(1.5, fam16)
b = FXnum(2.25, fam16)

print("a =", a)
print("b =", b)
print("a + b =", a + b)
print("a * b =", a * b)
print("sqrt(b) =", b.sqrt())
print("pi =", fam16.pi)
```
